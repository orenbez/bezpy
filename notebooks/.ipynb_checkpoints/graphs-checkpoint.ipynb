```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from datetime import datetime as dt
from math import pi
import numpy as np
```

```python
pi
```

```python
def f(x):
    return x**2 - 3*x 
```

```python
xlist_1 = np.linspace(-6, 9, 100)
```

```python
ylist_1 = f(xlist_1)
```

```python
plt.style.use('bmh')
plt.figure(num=0, dpi=128, figsize=(4, 3))
plt.plot(xlist_1, ylist_1, c='green')
plt.axis([-30, 30, -10, 60])
#plt.axis('scaled')   # draw graph with x-y to scale
plt.title("$y = x^2 - 3x$", fontsize=18)
plt.axhline(0, color='lightgray')   # x-axis line
plt.axvline(0, color='lightgray')  # y-axis line
plt.show()
```

```python
xlist_2 = np.arange(-2*pi, 2*pi, 0.1)    # alternative to the above with interval space
```

```python
len(xlist_2)
```

```python
ylist_2 = np.sin(xlist_2)
ylist_3 = np.cos(xlist_2)
ylist_4 = np.tan(xlist_2)
```

```python
plt.style.use('classic')
plt.figure(num=1, dpi=128, figsize=(8, 3))
plt.plot(xlist_2, ylist_2, c='red', label='sin(x)')
plt.plot(xlist_2, ylist_3, c='blue', label='cos(x)')
plt.axis([-pi, pi, -1.1, 1.1])
plt.fill_between(xlist_2, ylist_2, ylist_3, facecolor='pink', alpha=0.25)
plt.title("Trig Functions", fontsize=16)
plt.ylabel("y-axis", fontsize=14)
plt.xlabel("x-axis", fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=12)  # sets size of the labeled numbers on the axis
plt.legend()  # adds the labels
plt.axhline(0, color='lightgray')   # x-axis line
plt.axvline(0, color='lightgray')  # y-axis line

#plt.show()   # will not add the labels
```

```python
plt.style.use('classic')
plt.figure(num=1, dpi=128, figsize=(8, 3))
plt.plot(xlist_2, ylist_2, c='red', label='$\mathrm{sin}(x)$')
plt.plot(xlist_2, ylist_3, c='blue', label='$\mathrm{cos}(x)$')
plt.plot(xlist_2, ylist_4, c='green', label='$\mathrm{tan}(x)$')
plt.axis([-2 * pi, 2 * pi, -4, 4])
plt.fill_between(xlist_2, ylist_2, ylist_3, facecolor='pink', alpha=0.25)
plt.title("Trig Functions", fontsize=16)
plt.ylabel("y-axis", fontsize=14)
plt.xlabel("x-axis", fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=12)  # sets size of the labeled numbers on the axis
plt.legend()  # adds the labels
plt.axhline(0, color='lightgray')   # x-axis line
plt.axvline(0, color='lightgray')  # y-axis line

#plt.show()   # will not add the labels
```

```python
plt.style.use('classic')
plt.figure(num=1, dpi=128, figsize=(8, 3))

plt.plot(xlist_2, ylist_2, c='red', label='$\mathrm{sin}(x)$')
plt.plot(xlist_2, np.sin(pi/2 - xlist_2), c='blue', label='$\mathrm{sin}(90-x)$')
plt.plot(xlist_2, np.sin(xlist_2 - pi/2), c='green', label='$\mathrm{sin}(x-90)$')
plt.axis([-2 * pi, 2 * pi, -1.1, 1.1])

plt.title("Trig Functions", fontsize=16)
plt.ylabel("y-axis", fontsize=14)
plt.xlabel("x-axis", fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=12)  # sets size of the labeled numbers on the axis
plt.legend()  # adds the labels
plt.axhline(0, color='lightgray')   # x-axis line
plt.axvline(0, color='lightgray')  # y-axis line
```

```python
plt.style.use('classic')
plt.figure(num=1, dpi=128, figsize=(8, 3))
plt.plot(xlist_2, ylist_2, c='red', label='$\mathrm{sin}(x)$')
plt.plot(xlist_2, np.sin(-xlist_2), c='green', label='$\mathrm{sin}(-x)$')
plt.plot(xlist_2, np.sin(pi/2 - xlist_2), c='blue', label='$\mathrm{sin}(90-x)$')
#  Sin(-x + 90), this time the +90 is a shift to the right!!!
plt.axis([-2 * pi, 2 * pi, -1.1, 1.1])

plt.title("Trig Functions", fontsize=16)
plt.ylabel("y-axis", fontsize=14)
plt.xlabel("x-axis", fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=12)  # sets size of the labeled numbers on the axis
plt.legend()  # adds the labels
plt.axhline(0, color='lightgray')   # x-axis line
plt.axvline(0, color='lightgray')  # y-axis line
```

```python
plt.style.use('classic')
plt.figure(num=1, dpi=128, figsize=(8, 3))
plt.plot(xlist_2, ylist_2, c='red', label='$\mathrm{sin}(x)$')
plt.plot(xlist_2, 2*np.sin(xlist_2), c='blue', label='$2*\mathrm{sin}(x)$')
plt.plot(xlist_2, 3*np.sin(xlist_2), c='green', label='$3*\mathrm{sin}(x)$')
plt.axis([-2 * pi, 2 * pi, -3.1, 3.1])

plt.title("Trig Functions", fontsize=16)
plt.ylabel("y-axis", fontsize=14)
plt.xlabel("x-axis", fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=12)  # sets size of the labeled numbers on the axis
plt.legend()  # adds the labels
plt.axhline(0, color='lightgray')   # x-axis line
plt.axvline(0, color='lightgray')  # y-axis line
```

```python
plt.style.use('classic')
plt.figure(num=1, dpi=128, figsize=(8, 3))
xlist_2 = np.linspace(-10, 50, 100)
ylist_2 = np.exp2(xlist_2)

plt.plot(xlist_2, ylist_2, c='red', label='$\mathrm{2**}(x)$')
plt.plot(xlist_2, np.exp2(xlist_2-2), c='blue', label='$\mathrm{2**}(x-2)$')
# plt.plot(xlist_2, 3*np.sin(xlist_2), c='green', label='$3*\mathrm{sin}(x)$')
plt.axis([-6, 30, -5, 100])

plt.title("Exponential", fontsize=16)
plt.ylabel("y-axis", fontsize=14)
plt.xlabel("x-axis", fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=12)  # sets size of the labeled numbers on the axis
plt.legend()  # adds the labels
plt.axhline(0, color='lightgray')   # x-axis line
plt.axvline(0, color='lightgray')  # y-axis line
```
