```python
import networkx as nx                  #  requires `pip install networkx`
import matplotlib.pyplot as plt        #  requires `pip install matplotlib`
import pydot                           #  requires `pip install pydot`
```

```python
G = nx.DiGraph()   # empty graph object of Nodes(vertices) and Edges
G.add_edges_from([('A', 'B'), ('A', 'C'), ('C', 'B')])  # by default this is directional
```

```python
# choose a layout algorithm from here https://networkx.org/documentation/stable/reference/drawing.html#module-networkx.drawing.layout
pos = nx.spring_layout(G)
```

```python
# default settings
nx.draw_networkx(G, pos)
plt.show()
```

```python
# Customize
nx.draw_networkx_nodes(G, pos, node_size=100)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='red')
nx.draw_networkx_labels(G, pos)
plt.show()
```

```python
pos = nx.nx_pydot.graphviz_layout(G)
nx.draw_networkx(G, pos)
plt.savefig('networkx_graph.png')
```

```python
# Using pydot
pos = nx.nx_pydot.pydot_layout(G)
nx.draw_networkx(G, pos)
```
